Boxplot
Summary Statistics: Shows the median, quartiles, and potential outliers.

Structure: Consists of a box (representing the interquartile range), a line inside the box (the median), and whiskers that extend to the minimum and maximum values within 1.5 times the interquartile range. Outliers are plotted as individual points.

Violin Plot
Summary Statistics and Density: Combines the features of a boxplot with a kernel density plot, showing the distributions shape and density.

Structure: Includes a boxplot within a violin shape that represents the datas density. The width of the violin at different values indicates the density of the data at those values.


Use Case: Useful for understanding the distributions shape, especially when comparing multiple distributions.
