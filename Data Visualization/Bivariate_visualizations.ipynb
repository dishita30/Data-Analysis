## BIVARIATE VISUALIZATIONS(CATEGORICAL)

```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
sns.countplot(x = "survived",data=df, hue = "class")
```

```python
x = pd.crosstab(df["survived"], df["class"])
```

### PIVOT TABLE

```python
x
```

```python
sns.heatmap(x,annot=True,fmt = ".1f")
```

```python
sns.heatmap(x,annot=True,fmt = ".1f",cmap="coolwarm")
```

color changed by cmap attribute 

```python
sns.stripplot(x = "class", y = "age", hue ="sex",data = df)
```
