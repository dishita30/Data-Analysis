```python
import pandas as pd 
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
```

```python
df.head()
```

```python
sns.barplot(x = "class", y = "age", data = df)
```

```python
sns.barplot(x = "class", y = "age", data = df,estimator="sum")
```

accumulation of the sum of all people travelling in first class, and so on

```python
sns.barplot(x = "class", y = "fare", data = df,estimator="sum")
```

```python
sns.barplot(x = "class", y = "age", hue = "sex",data = df)
```

first second and third class are further categorized into males and females

```python
sns.boxplot(x = "class", y = "age",hue = "survived",data = df)
```

```python
sns.violinplot(x = "class", y = "age",hue = "survived",data = df)
```
