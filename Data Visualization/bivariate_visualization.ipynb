### LOADING ANOTHER DATASET

```python
df2 = sns.load_dataset("tips")
df2.head()
```

```python
sns.scatterplot(x="total_bill", y ="tip",hue = "sex",style ="smoker",data = df2)
```

different style for smokers


## Joint Plot


```python
sns.jointplot(x="total_bill", y = "tip",data = df2)
```

```python
sns.jointplot(x="total_bill", y = "tip",hue = "sex",data = df2)
```

instead  of histogram, joint probability distribution of males and females since two categories and not one

```python
sns.jointplot(x="total_bill", y = "tip",data = df2,kind = "reg")
```

to get a line on a scatter plot in ml, best fit for these values
