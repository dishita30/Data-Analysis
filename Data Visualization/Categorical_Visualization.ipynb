## UNIVARIATE CATEGORICAL VISUALIZATION

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('AB_NYC_2019.csv')
df.head(2)
```

### Categorical

## Bar Graph

```python
sns.countplot(data = df , x = "neighbourhood_group")
```

```python
sns.countplot(data = df , x = "room_type")
```

```python
sns.countplot(data = df , x = "neighbourhood_group", hue = "room_type")
```

room types count in per neighbourhood group

## Pie Chart

```python
df["room_type"].value_counts().plot(kind="pie",autopct = "%.2f")
```

auto pct shows percentage value which is fixed till two decimal digits in this



```python
df["neighbourhood_group"].value_counts().plot(kind="pie",autopct = "%.2f")
```
